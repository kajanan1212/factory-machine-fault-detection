```python
import os
import pandas as pd
```

```python
vibration_data = []
machine_status = []

data_list = os.listdir("./data/xep-vibration-data")

for w in data_list:
    if "week" in w:
        week_data_list = os.listdir(f"./data/xep-vibration-data/{w}")
        for d in week_data_list:
            if "vibrationdata" in d:
                vibration_data.append(f"./data/xep-vibration-data/{w}/{d}")
            elif "machinestatus" in d:
                machine_status.append(f"./data/xep-vibration-data/{w}/{d}")
```

```python
vibration_df = pd.concat([pd.read_csv(v) for v in vibration_data], ignore_index=True)
machine_status_df = pd.concat([pd.read_csv(m) for m in machine_status], ignore_index=True)
```

```python
sensor1 = machine_status_df.loc[machine_status_df["UID"] == "sensor1"]
sensor1.isnull().sum()
```

```python
sensor2 = machine_status_df.loc[machine_status_df["UID"] == "sensor2"]
sensor2.isnull().sum()
```

```python
sensor3 = machine_status_df.loc[machine_status_df["UID"] == "sensor3"]
sensor3.isnull().sum()
```

```python
sensor1 = sensor1.dropna(axis=1).drop(columns=["UID"])
sensor2 = sensor2.dropna(axis=1).drop(columns=["UID"])
sensor3 = sensor3.dropna(axis=1).drop(columns=["UID"])

sensor1["sensor1_Timestamp"] = sensor1["Timestamp"]
sensor2["sensor2_Timestamp"] = sensor2["Timestamp"]
sensor3["sensor3_Timestamp"] = sensor3["Timestamp"]
```

```python
status_mapped_df = pd.merge_asof(
    pd.merge_asof(
        sensor1, sensor2, on="Timestamp", direction="nearest"
    ),
    sensor3, on="Timestamp", direction="nearest"
)

status_mapped_df
```

```python
status_mapped_df["s1_Timestamp_minutes_diff"] = (status_mapped_df["Timestamp"] - status_mapped_df["sensor1_Timestamp"]).abs() / (1000 * 60)
status_mapped_df["s2_Timestamp_minutes_diff"] = (status_mapped_df["Timestamp"] - status_mapped_df["sensor2_Timestamp"]).abs() / (1000 * 60)
status_mapped_df["s3_Timestamp_minutes_diff"] = (status_mapped_df["Timestamp"] - status_mapped_df["sensor3_Timestamp"]).abs() / (1000 * 60)
```

```python
status_mapped_df[[
    "Timestamp", "s1_Timestamp_minutes_diff", "s2_Timestamp_minutes_diff", "s3_Timestamp_minutes_diff"
]].describe()
```

```python
tolerance = 1
tolerance
```

```python
status_mapped_df.loc[
    (status_mapped_df["s1_Timestamp_minutes_diff"] > tolerance) |
    (status_mapped_df["s2_Timestamp_minutes_diff"] > tolerance) |
    (status_mapped_df["s3_Timestamp_minutes_diff"] > tolerance)
]
```

```python
s1_m = [f"machine{i}" for i in range(1, 6)]
s2_m = [f"machine{i}" for i in range(6, 14)]
s3_m = [f"machine{i}" for i in range(14, 22)]

def handle_outlier_machine_status(row):
    if row["s1_Timestamp_minutes_diff"] > tolerance:
        row[s1_m] = 0
    if row["s2_Timestamp_minutes_diff"] > tolerance:
        row[s2_m] = 0
    if row["s3_Timestamp_minutes_diff"] > tolerance:
        row[s3_m] = 0

    return row
```

```python
status_mapped_df = status_mapped_df.apply(
    handle_outlier_machine_status,
    axis=1,
)
```

```python
status_mapped_df = status_mapped_df.drop(columns=[
    "sensor1_Timestamp", "sensor2_Timestamp", "sensor3_Timestamp",
    "s1_Timestamp_minutes_diff", "s2_Timestamp_minutes_diff", "s3_Timestamp_minutes_diff",
])
```

```python
mapped_df = vibration_df.merge(status_mapped_df, on=["Timestamp"], how="inner")
mapped_df
```

```python
vibration_df["Timestamp"] = vibration_df["Timestamp"] / (1000 * 60)
vibration_df["Timestamp"] = vibration_df["Timestamp"].astype("int64")
vibration_df["DateTime"] = pd.to_datetime(vibration_df["Timestamp"], unit="m")
vibration_df = vibration_df.sort_values("DateTime", ignore_index=True)

status_mapped_df["Timestamp"] = status_mapped_df["Timestamp"] / (1000 * 60)
status_mapped_df["Timestamp"] = status_mapped_df["Timestamp"].astype("int64")
status_mapped_df["DateTime"] = pd.to_datetime(status_mapped_df["Timestamp"], unit="m")
status_mapped_df = status_mapped_df.sort_values("DateTime", ignore_index=True)
```

```python
front_cols = ["Timestamp", "DateTime"]
vibration_df = vibration_df[front_cols + [col for col in vibration_df if col not in front_cols]]
status_mapped_df = status_mapped_df[front_cols + [col for col in status_mapped_df if col not in front_cols]]
```

```python
mapped_df = vibration_df.merge(status_mapped_df, on=["Timestamp", "DateTime"], how="inner")
mapped_df = mapped_df.sort_values(["Timestamp", "DateTime"], ignore_index=True)
mapped_df
```
