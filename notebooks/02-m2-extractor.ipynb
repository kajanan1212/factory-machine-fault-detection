```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mapped_df = pd.read_csv("./data/mapped.csv")
mapped_df
```

```python
pure_noisy_m2 = mapped_df.drop(mapped_df[mapped_df["machine2"] != 1].index).reset_index(drop=True)
pure_noisy_m2
```

```python
pure_noisy_m2.info()
```

```python
all_m_ex_m2_status = pure_noisy_m2.loc[:, [f"machine{i}" for i in range(1, 6)]].drop(columns=["machine2"])
all_m_ex_m2_status
```

```python
(all_m_ex_m2_status.eq(1) | all_m_ex_m2_status.eq(-1)).any(axis=1).value_counts()
```

```python
all_m_ex_m2_status.eq(1).any(axis=1).value_counts()
```

```python
is_noisy_m2 = (all_m_ex_m2_status.eq(1) | all_m_ex_m2_status.eq(-1)).any(axis=1)
pure_noisy_m2["is_pure"] = is_noisy_m2.apply(lambda is_noisy: 0 if is_noisy else 1)
pure_noisy_m2["is_pure"].value_counts()
```

```python
sns.countplot(pure_noisy_m2["is_pure"])
```

```python
pure_noisy_m2 = pure_noisy_m2.drop(columns=[f"machine{i}" for i in range(1, 22)])
```

```python
try:
    pure_noisy_m2 = pure_noisy_m2.sort_values("Timestamp", ignore_index=True)
except KeyError:
    pure_noisy_m2 = pure_noisy_m2.sort_values(["converted", "Timestamp_v", "Timestamp_s"], ignore_index=True)
```

```python
pure_noisy_m2.to_csv("./data/pure_noisy_m2.csv", index=False)
```
