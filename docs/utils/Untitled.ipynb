```python
machine_status_df = machine_status_df.drop_duplicates(["Timestamp", "UID"])
```

```python
sensor1 = machine_status_df.loc[machine_status_df["UID"] == "sensor1"]
sensor1.isnull().sum()
```

```python
sensor2 = machine_status_df.loc[machine_status_df["UID"] == "sensor2"]
sensor2.isnull().sum()
```

```python
sensor3 = machine_status_df.loc[machine_status_df["UID"] == "sensor3"]
sensor3.isnull().sum()
```

```python
sensor1 = sensor1.dropna(axis=1)
sensor2 = sensor2.dropna(axis=1)
sensor3 = sensor3.dropna(axis=1)
```

```python
sensor1
```

```python
sensor1.loc[sensor1["machine2"] == 1, "machine5"].value_counts()
```

```python
pure_m2 = sensor1.loc[
    (sensor1["machine1"] == 0) &
    (sensor1["machine2"] == 1) &
    (sensor1["machine3"] == 0) &
    (sensor1["machine4"] == 0) &
    (sensor1["machine5"] == 0)
]
pure_m2["is_pure_m2"] = 1
pure_m2 = pure_m2[["Timestamp", "is_pure_m2"]]
pure_m2
```

```python
noisy_m2 = sensor1.loc[
    (sensor1["machine2"] == 1) &
    (
        (sensor1["machine1"] != 0) |
        (sensor1["machine3"] != 0) |
        (sensor1["machine4"] != 0) |
        (sensor1["machine5"] != 0)
    )
]
noisy_m2["is_pure_m2"] = 0
noisy_m2 = noisy_m2[["Timestamp", "is_pure_m2"]]
noisy_m2
```

```python
absent_m2 = sensor1.loc[
    (sensor1["machine2"] != 1)
]
absent_m2["is_pure_m2"] = -1
absent_m2 = absent_m2[["Timestamp", "is_pure_m2"]]
absent_m2
```

```python
m2_status_df = pd.concat([pure_m2, noisy_m2, absent_m2]).sort_values("Timestamp", ignore_index=True)
m2_status_df
```
